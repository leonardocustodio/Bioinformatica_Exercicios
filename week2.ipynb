### Encontra a posição dos menores Skews em uma sequência

```python
%%time
def lower_skews_positions(sequence):
    rules = {
        "G": 1,
        "C": -1,
        "A": 0,
        "T": 0,
    }
    
    min_skew_value = 0
    min_skew_indexes = []
    
    skew_value = 0
    skews = "0"
    for n in range(len(sequence)):
        skew_value += rules[sequence[n]]
        skews += " " + str(skew_value)
        
        if (skew_value == min_skew_value):
            min_skew_indexes.append(n+1)
        elif (skew_value < min_skew_value):
            min_skew_value = skew_value
            min_skew_indexes.clear()
            min_skew_indexes.append(n+1)

        
    print("Min Value: " + str(min_skew_value))
    return min_skew_indexes


sequence = ""
lower_skews_positions(sequence)
    
```
