### Gerar "vizinha" de k-mers que são similares (considerado iguais se houver apenas d mismatches)

Neighbors(CAA, 1)

Tira a primeira letra, fica "AA"

Encontra os que tem uma hamming distance de até D

AA, CA, GA, TA, AC, AG, AT

Adiciona o C na frente, dai fica:

res1: CAA, CCA, CGA, CTA, CAC, CAG, CAT

Troca a primeira letra de CAA por qualquer um dos 4 nucleotídeos

res2: AAA, CAA, GAA, TAA


Faz um set de res1 + res2:

CAA, CCA, CGA, CTA, CAC, CAG, CAT, AAA, GAA, TAA

Da um total de 10.

```python
def make_neighborhood_set(pattern,d):
    bases = {'A', 'T', 'C', 'G'}
    
    def neighbours(p):
        neighbours=[]
        for i in range(len(p)):
            for ch in bases:
                neighbours.append(p[0:i]+ch+p[i+1:])
        return neighbours
    if d==0:
        return [pattern]
    else:
        neighbourhood=[]
        start=make_neighborhood_set(pattern,d-1)
        for p in start:
            for n in neighbours(p):
                neighbourhood.append(n)
        return list(set(start+neighbourhood))
        
        

    
pattern = 'CAA'
hamming_distance = 1
make_neighborhood_set(pattern, hamming_distance)
```

        for j ← 0 to |neighborhood| - 1
        
            neighbor ← neighborhood[j]
            
            if freqMap[neighbor] doesn't exist
            
                freqMap[neighbor] ← 1
                
            else
            
                freqMap[neighbor] ← freqMap[neighbor] + 1
                
    m ← MaxMap(freqMap)
    
    for every key Pattern in freqMap
    
        if freqMap[Pattern] = m
        
            append Pattern to Patterns
            
    return Patterns

```python
def find_frequent_words_with_mismatches(sequence, pattern_size, mismatches):
    patterns = []
    freq = {}
    
    for i in range(len(sequence) - pattern_size):
        pattern = sequence[i:i+pattern_size]
        neighborhood = make_neighborhood_set(pattern, mismatches)
        for x in range(len(neighborhood)):
            neighbor = neighborhood[x]
            if (neighbor in freq):
                freq[neighbor] = freq[neighbor] + 1
            else:
                freq[neighbor] = 1
                
    #print(freq)
    max_freq = max(freq.values())
    most_frequent = []
    for y in freq.keys():
        if freq[y] == max_freq:
            most_frequent.append(y)
        
    return most_frequent
    
    
sequence = 'GCGTTAGGCGTACACGCTTGCGTTTGAGCGTGCTTTAGACACTTGAGCGTGCGTTTGAGCTTGCTTGCGTACACGCGTGCTTGCTTGCTTACACTTGAGCTTGCTTGCGTGCTTTAGGCTTACACGCTTTTGAGCTTTTGAGCTTTTGAGCGTTAGGCGTTTGAGCTTTAGTAGGCTTACACTTGAGCGTTAGTAGTAGTTGAGCTTTAGTAGTAGTAGTTGATTGATTGAGCTTTAGGCTTTAGTTGATTGAACACGCTTGCTTTAGGCGTGCTTGCGTGCGTACACTTGAGCGTACACGCGTTTGATTGATAGTAGGCTTTTGAGCTTGCGTGCTTGCGTACACTTGAGCGTACACGCGTACACTAGTAGGCGT'
pattern_size = 5
mismatches = 3

find_frequent_words_with_mismatches(sequence, pattern_size, mismatches)
```
